## Map

```python
import math
```

```python
def area(r):
    return math.pi * (r**2)
```

```python
radii = [2, 5, 7.1, 0.3, 10]
```

The "classic" approach:

```python
areas=[]
for i in radii:
    areas.append(area(i))
print(areas)
```

The "pythonic" approach, using map():

```python
c = list(map(area, radii))
print(c)
```

## Filter

```python
import statistics

data = [1.5, 9.2, 3.4, 5.1, -9,4]

avg = statistics.mean(data)
```

The classic approach:

```python
def above_avg(avg, number):
    if number <= avg:
        return False
    else:
        return True
```

```python
above_avg(avg=avg, number = 4)
```

The "pythonic" approach, using filter():

```python
def above_avg(i):
    return i>avg

list(filter(above_avg, data))
```

```python
above_avg(4)
```

## Reduce

```python
from functools import reduce
```

```python
names = ["Guillem", "Javi", "Mar", "Pol"]
```

```python
def concatenator(x, y):
    return x + " " + y
```

```python
reduce(concatenator, names)
```

```python
numbers = [1,2,3,4,5]
```

Alternative approaches:

```python
s = ""
for e in names:
    s += e + " "
print(s)
```
